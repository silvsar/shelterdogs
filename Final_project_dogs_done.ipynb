```python
all_animals = []  # List to store all dogs
page = 1 # Start with page 1 
limit = 40  # Dogs per page

while True: # Start an infinite loop 
    params = {"type": "dog", "location": "New York", "page": page, "limit": limit} # Set parameters 
    response = requests.get(url, headers=headers, params=params) # Sends request to API 
    data = response.json() # Converts the APIâ€™s JSON response into a Python dictionary.
    animals = data.get('animals', [])  # Dogs from this page
    all_animals.extend(animals)        # Add to main list
    if len(animals) < limit:           # No more pages left
        break
    page += 1

print(len(all_animals))  # Print total number of dogs
```
